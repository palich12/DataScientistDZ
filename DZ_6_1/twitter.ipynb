## 2. Строим частотный список

```python
import re
import collections

def get_freq_list(texts):
    words = collections.defaultdict(int)
    for text in texts:
        for word in re.split(r'\W', text):
            words[word] += 1
    del words[u'']
    
    return [(word, words[word]) for word in sorted(words, key=lambda w: words[w], reverse=True)]
```

```python
get_freq_list(x_train['ttext'])
```

```python
import sklearn.pipeline
import sklearn.naive_bayes

text_clf = sklearn.pipeline.Pipeline([('vect', sklearn.feature_extraction.text.CountVectorizer()),
                     ('clf', sklearn.naive_bayes.MultinomialNB())])
```

```python
text_clf
```

## 4. Classifying

```python
text_clf.fit(x_train["ttext"], y_train)
```

```python
predicted = text_clf.predict(x_val["ttext"])
```

```python
import sklearn.metrics
print(sklearn.metrics.classification_report(y_val, predicted, target_names=["negative", "positive"]))
```

```python
sklearn.metrics.confusion_matrix(y_val, predicted)
```
